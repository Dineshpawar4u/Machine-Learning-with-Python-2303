## Method Used

"K Means Clustering" is an unsupervised learning algorithm that tries to cluster data according to their similarity. Unsupervised learning means there are no results to predict and the algorithm is just trying to find patterns in the data. In the phrase "k means clustering" we specified the number of clusters we want the data to be grouped into. The algorithm assigns each observation to a random cluster and finds the center of each cluster. Next, the algorithm is iterated in two steps:
Reassign data points to the cluster closest to the centroid. Calculate the new center of gravity for each cluster. These two steps are repeated until the intra-cluster variation cannot be further reduced. Intra-cluster variation is calculated as the sum of the euclidean distance between the data points and the respective cluster weights.

## Episode 1: Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Episode 2: Create Data

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

## Episode 3: Visualize Data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## Episode 4: Creating the Clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
